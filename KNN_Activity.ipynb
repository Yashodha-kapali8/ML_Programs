**ACTIVITY:**

Develop a program to implement k-Nearest Neighbour algorithm to classify the randomly generated 100 values of X in the range of [0 1]. Perform the following based on dataset generated" a. Label the first 50 points {x1, ..., x50} as follows; if xi <= 0.5 then xi blongs to class C1 else class C2. b. Classify the remaining points {x51, ..., x100} using KNN. Perform this for k = 1,2,3,4,5,20,30.

```
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Generate 100 random values in [0, 1]
np.random.seed(42)  # for reproducibility
X = np.random.rand(100, 1)

# Label the first 50 values
y = np.array(["C1" if x <= 0.5 else "C2" for x in X[:50].flatten()])
```

```
# Split data into training and test sets
X_train = X[:50]
y_train = y
X_test = X[50:]
```

```
#  Test for multiple values of k
k_values = [1, 2, 3, 4, 5, 20, 30]

print("KNN Classification Results:\n")
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    print(f"k = {k}")
    for i in range(len(X_test)):
        print(f"Test x = {X_test[i][0]:.3f} --> Predicted Class = {y_pred[i]}")
    print("-" * 40)

# Visualize classification for k=3
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

```

```
plt.figure(figsize=(10, 3))
plt.scatter(X_train, [1]*50, c=['blue' if c == 'C1' else 'red' for c in y_train], label='Train')
plt.scatter(X_test, [2]*50, c=['blue' if c == 'C1' else 'red' for c in y_pred], marker='x', label='Test (Predicted)')
plt.yticks([1, 2], ['Train', 'Test'])
plt.xlabel("X values")
plt.title("KNN Classification (k=3)")
plt.legend()
plt.grid(True)
plt.show()
```
