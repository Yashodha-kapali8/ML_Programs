**LINEAR REGRESSION:**


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

#  Load the CSV thatâ€™s already present
df = pd.read_csv('testdata.csv')   # make sure the filename matches exactly

# Split into input (Temperature) and output (Pressure)
temperature = df.iloc[:, 1:2].values   # assumes temperature is the 2nd column
pressure    = df.iloc[:, 2].values     # assumes pressure    is the 3rd column

# Fit simple linear-regression model
linreg = LinearRegression()
linreg.fit(temperature, pressure)

#  Predictions
predicted = linreg.predict(temperature)
```

```
# Plot
plt.figure(figsize=(8, 5))
plt.scatter(temperature, pressure,  color='red',  label='Real Data')
plt.scatter(temperature, predicted, color='blue', label='Predicted')
plt.title('Temperature vs Pressure (Linear Regression)')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.show()
```

```
# Model details and error
print(f"Coefficient (slope): {linreg.coef_[0]:.4f}")
print(f"Intercept: {linreg.intercept_:.4f}")
mae = mean_absolute_error(pressure, predicted)
print(f"Mean Absolute Error (MAE): {mae:.4f}")

# Predict pressure for a new temperature (e.g., 30 Â°C)
new_temp = np.array([[30]])
pred_pressure = linreg.predict(new_temp)[0]
print(f"Predicted pressure for temperature {new_temp[0][0]} = {pred_pressure:.4f}")

#  Same prediction using the explicit formula
pred_pressure_formula = linreg.coef_[0] * new_temp + linreg.intercept_
print(f"Computed via formula: {pred_pressure_formula[0][0]:.4f}")
```
