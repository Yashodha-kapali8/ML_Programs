**POLYNOMIAL REGRESSION:**



```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```
# Load dataset from file already present in Colab
df = pd.read_csv('testdata.csv')
```

```
# Extract input (Temperature) and output (Pressure)
temperature = df.iloc[:, 1:2].values
pressure = df.iloc[:, 2].values
```

```
# Create polynomial features (degree = 3)
polyreg = PolynomialFeatures(degree=3)
X_poly = polyreg.fit_transform(temperature)
# Fit Polynomial Regression model
linreg = LinearRegression()
linreg.fit(X_poly, pressure)

# Predict pressure values using the polynomial model
predicted_pressure = linreg.predict(X_poly)

```

```
# Visualize the results
plt.figure(figsize=(8, 5))
plt.scatter(temperature, pressure, color='red', label='Actual')
plt.plot(temperature, predicted_pressure, color='blue', label='Predicted')
plt.title('Temperature vs Pressure (Polynomial Regression)')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.show()
```
